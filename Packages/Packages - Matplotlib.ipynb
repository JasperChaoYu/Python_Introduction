## Matplotlib 是在 Python 上以 matlab 圖形為基底的畫圖模組。

```python
import matplotlib
from matplotlib import pyplot as plt
```

```python
A = [1,2,3,4,5]
B = [1,2,3,4,5]
plt.plot(A, B)
plt.show()
```

```python
% matplotlib inline
A = [1,2,3,4,5]
B = [1,2,3,4,5]
plt.plot(A, B)
```

```python
A = [1,2,3,4,5]
B = [1,2,3,4,5]
plt.plot(A, B)
plt.grid()
plt.show()
```

```python
A = [1,2,3,4,5]
B = [1,2,3,4,5]
plt.plot(A, B)
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
A = [1,2,3,4,5]
B = [1,2,3,4,5]
plt.plot(A, B)
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["result"],loc="upper left")
'''
# valid locations are:
right
center left
upper right
lower right
best
center
lower left
center right
upper left
upper center
lower center
'''
plt.title("X=Y")
plt.show()
```

```python
A = [1,2,3,4,5]
B = [1,2,3,4,5]
plt.plot(A, B)
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["result"],loc="upper left")
plt.title("X=Y")
plt.show()
```

```python
from math import pi
A = [0.5*pi,1*pi,1.5*pi,2*pi]
B = [1,2,3,4]
plt.polar()
plt.plot(A,B)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.pie([1,2,3,4,5], labels=[1,2,3,4,5], shadow=True, frame=True)

plt.subplot(2,2,2)
plt.hist([11,2,4,5,2112,4,5,46,45,76,745,3245,2,54])

#subplot(nrows, ncols, plot_number)
```

```python
import numpy
x = numpy.linspace(0,5,11)
y = x**2
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x,'r')
plt.show()
```
